## NOMOR 2 a)

```python
import numpy as np
import matplotlib.pyplot as plt
# Sampling rate
fs = 10000
# Waktu
t = np.arange(0, 1, 1/fs)
# Sinyal analog
x = 4 * np.cos(3 * np.pi * t) + 6 * np.sin(2 * np.pi * t)
# Plot sinyal analog
plt.plot(t, x)
plt.xlabel('Waktu (t)')
plt.ylabel('Amplitudo')
plt.title('Bentuk Pulsa Gelombang x(t)')
plt.show()
```

## NOMOR 2 b)

```python
# Quantization level
L = 1024
# Quantized signal
x_q = np.round((x + 1) * (L / 2)) - 1
# Plot sinyal diskrit
plt.stem(t, x_q)
plt.xlabel('Waktu (t)')
plt.ylabel('Amplitudo Terkuantisasi')
plt.title('Bentuk Pulsa Gelombang Diskrit x(t)')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.fft import fft

# Define the function f(x) = x^2
def f(x):
    return x**2

# Define the interval 0 < x < 2π
x_values = np.linspace(0, 2*np.pi, 100)

# Calculate the function values
y_values = f(x_values)

# Plot f(x)
plt.figure(figsize=(8, 4))
plt.plot(x_values, y_values, label="f(x) = x^2")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("Plot of f(x) = x^2")
plt.grid(True)
plt.legend()
plt.show()

# Calculate the Fourier coefficients analytically
def fourier_coefficients(n):
    a0 = (1/(2*np.pi)) * np.trapz(y_values, x_values)
    an = (1/np.pi) * np.trapz(y_values * np.cos(n*x_values), x_values)
    bn = (1/np.pi) * np.trapz(y_values * np.sin(n*x_values), x_values)
    return a0, an, bn

# Calculate the Fourier series
def fourier_series(x, n):
    a0, an, bn = fourier_coefficients(n)
    series = a0 + np.sum(an * np.cos(n*x) + bn * np.sin(n*x))
    return series

# Calculate the Fourier series for n=100
n_values = np.arange(1, 101)
fourier_values = np.array([fourier_series(x_values, n) for n in n_values])

# Plot the Fourier series
plt.figure(figsize=(8, 4))
plt.plot(x_values, y_values, label="f(x) = x^2")
plt.plot(x_values, fourier_values, label="Fourier series (n=100)")
plt.xlabel("x")
plt.ylabel("Value")
plt.title("Fourier Series of f(x) = x^2")
plt.grid(True)
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the function x(n) = 1/n * u(n-1)
def x(n):
    return 1 / n if n > 0 else 0

# Define the range of n
n_values = np.arange(0, 20)

# Calculate the Z-transform of x(n)
def z_transform(n_values):
    X_z = np.array([x(n) for n in n_values])
    return X_z

# Plot the magnitude of X(z)
plt.figure(figsize=(8, 4))
plt.stem(n_values, np.abs(z_transform(n_values)), basefmt="b-")
plt.xlabel("n")
plt.ylabel("|X(z)|")
plt.title("Magnitude of X(z) for x(n) = 1/n * u(n-1)")
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt 
from scipy.signal import freqz

#Contoh sinyal atau urutan 
sinyal = np.array([1, 2, 3, 4, 5])
# Koefisien penyebut dan pembilang (untuk contoh sederhana)
a = np.array([1, 1])
# koefisien penyebut (denominator)
b = sinyal
#koefisien pembilang (numerator)
#Menghitung respon frekuensi
w, h = freqz(b, a)
#Plot respon magnitudo
plt.figure(figsize=(12, 6))
plt.subplot(2, 1, 1)
plt.plot(w, np.abs(h))
plt.title('Respon Magnitudo')
plt.ylabel('Magnitudo')
plt.xlabel('Frekuensi [rad/sample]')
plt.grid()

# Plot respon fase
plt.subplot(2, 1, 2)
plt.plot(w, np.angle(h))
plt.title('Respon Fase')
plt.ylabel("Fase (radlan)")
plt.xlabel('Frekuensi [rad/sample]')
plt.grid()
plt.tight_layout()
plt.show()
```
