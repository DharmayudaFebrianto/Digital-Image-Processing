## Import Libraries package

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

## Mengekstrak komponen warna red green dan blue

```python
# Load citra warna (gantikan dengan nama file citra warna yang sebenarnya)
image_path = "Singa.jpeg"
image = cv2.imread(image_path, cv2.IMREAD_COLOR)

# Ekstrak komponen merah, hijau, dan biru
blue_channel = image[:, :, 0]
green_channel = image[:, :, 1]
red_channel = image[:, :, 2]

# Tampilkan citra asli dan komponen-komponennya
fig, axes = plt.subplots(1, 4, figsize=(16, 4))

axes[0].imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
axes[0].set_title("Citra Asli")
axes[0].axis("off")

axes[1].imshow(blue_channel, cmap="gray")
axes[1].set_title("Komponen Biru")
axes[1].axis("off")

axes[2].imshow(green_channel, cmap="gray")
axes[2].set_title("Komponen Hijau")
axes[2].axis("off")

axes[3].imshow(red_channel, cmap="gray")
axes[3].set_title("Komponen Merah")
axes[3].axis("off")

plt.tight_layout()
plt.show()
```

## Menampilkan Histogram dari Citra yang digunakan

```python
# Load the image (assuming you have an image file named "image.jpg")
image_path = "Singa.jpeg"
image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

# Calculate histogram
hist, bins = np.histogram(image, bins=256, range=[0, 256])

# Calculate total number of pixels
total_pixels = image.shape[0] * image.shape[1]

# Plot the histogram
plt.figure(figsize=(10, 6))
plt.bar(bins[:-1], hist, width=1, color='b', alpha=0.7)
plt.title("Histogram of Grayscale Image")
plt.xlabel("Pixel Intensity")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()

# Print the results
print(f"Total number of pixels: {total_pixels}")
print(f"Number of pixels for each intensity level:")
for i, count in enumerate(hist):
    print(f"Intensity {i}: {count} pixels")
```
