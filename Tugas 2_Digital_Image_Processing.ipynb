## Simple Thresholding

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Read an example grayscale image (you can replace 'gradient.png' with your own image)
img = cv.imread('Gambar Tugas 2.jpg', cv.IMREAD_GRAYSCALE)
assert img is not None, "File could not be read, check with os.path.exists()"
```

```python
# Apply different types of simple thresholding
ret, thresh1 = cv.threshold(img, 127, 255, cv.THRESH_BINARY)
ret, thresh2 = cv.threshold(img, 127, 255, cv.THRESH_BINARY_INV)
ret, thresh3 = cv.threshold(img, 127, 255, cv.THRESH_TRUNC)
ret, thresh4 = cv.threshold(img, 127, 255, cv.THRESH_TOZERO)
ret, thresh5 = cv.threshold(img, 127, 255, cv.THRESH_TOZERO_INV)
```

```python
# Display the results
titles = ['Original Image', 'BINARY', 'BINARY_INV', 'TRUNC', 'TOZERO', 'TOZERO_INV']
images = [img, thresh1, thresh2, thresh3, thresh4, thresh5]
```

```python
for i in range(6):
    plt.subplot(2, 3, i + 1)
    plt.imshow(images[i], 'gray', vmin=0, vmax=255)
    plt.title(titles[i])
    plt.xticks([]), plt.yticks([])

plt.show()
```

## Global Thresholding

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

# Load citra ECG
ecg_image = cv.imread('Gambar Tugas 2.jpg', cv.IMREAD_GRAYSCALE)

# Global thresholding
_, global_thresholded = cv.threshold(ecg_image, 127, 255, cv.THRESH_BINARY)

# Display the results
plt.figure(figsize=(8, 4))
plt.subplot(121), plt.imshow(ecg_image, cmap="gray"), plt.title("Original ECG Image")
plt.subplot(122), plt.imshow(global_thresholded, cmap="gray"), plt.title("Global Thresholding")
plt.tight_layout()
plt.show()
```

## Local Adaptive Thresholding

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

# Local Adaptive Thresholding level 1
adaptive_thresholded = cv.adaptiveThreshold(ecg_image, 255, cv.ADAPTIVE_THRESH_MEAN_C, cv.THRESH_BINARY, 11, 2)

# Display the results
plt.figure(figsize=(8, 4))
plt.subplot(121), plt.imshow(ecg_image, cmap="gray"), plt.title("Original ECG Image")
plt.subplot(122), plt.imshow(adaptive_thresholded, cmap="gray"), plt.title("Local Adaptive Thresholding (level 1)")
plt.tight_layout()
plt.show()
```

## Adaptive Thresholding

```python
# Load the ECG image (replace with the actual ECG image file)
ecg_image = cv.imread("Gambar Tugas 2.jpg", cv.IMREAD_GRAYSCALE)

# Apply median blur
img = cv.medianBlur(ecg_image, 5)

# Global thresholding
_, th1 = cv.threshold(img, 150, 255, cv.THRESH_BINARY)

# Adaptive mean thresholding
th2 = cv.adaptiveThreshold(img, 255, cv.ADAPTIVE_THRESH_MEAN_C,
                            cv.THRESH_BINARY, 11, 2)

# Adaptive Gaussian thresholding
th3 = cv.adaptiveThreshold(img, 255, cv.ADAPTIVE_THRESH_GAUSSIAN_C,
                            cv.THRESH_BINARY, 11, 2)

# Display the results
titles = ['Original Image', 'Global Thresholding (v = 150)',
          'Adaptive Mean Thresholding', 'Adaptive Gaussian Thresholding']
images = [img, th1, th2, th3]

plt.figure(figsize=(10, 6))
for i in range(4):
    plt.subplot(2, 2, i + 1)
    plt.imshow(images[i], 'gray')
    plt.title(titles[i])
    plt.xticks([]), plt.yticks([])

plt.tight_layout()
plt.show()
```
